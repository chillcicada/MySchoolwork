```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 1000
plt.rcParams['font.sans-serif'] = ['LXGW WenKai']
```

```python
# 获取相对误差
def get_rel_error(data):
    arg = np.mean(data)
    max_data = np.max(data)
    min_data = np.min(data)
    abs_error = max(max_data - arg, arg - min_data)
    rel_error = abs_error / arg
    print(f'相对误差 = {rel_error}')
```

```python
"""
丙烯酸质量 = 5g
中和度 = 50%
引发剂比例 = 1.6%
index = 1 -> 7

plot: 交联剂比例-吸水倍率
"""

丙烯酸质量 = [
    5.018,
    5.047,
    5.02,
    5,
    5.009,
    4.997,
    4.994,
]

get_rel_error(丙烯酸质量)

引发剂比例 = [
    0.016739737,
    0.016445413,
    0.015936255,
    0.016,
    0.016570174,
    0.015609366,
    0.016019223,
]

get_rel_error(引发剂比例)

中和度 = [
    50.265,
    49.976,
    50.245,
    50.446,
    50.355,
    50.476,
    50.506,
]

get_rel_error(中和度)

# 交联剂比例-吸水倍率 plot

交联剂比例 = [
    0.007971303,
    0.016049138,
    0.023705179,
    0.0402,
    0.061090038,
    0.079847909,
    0.099919904,
]

吸水倍率 = [
    202.952381,
    178.3823529,
    98.1509434,
    98.66990291,
    64.30392157,
    48.40566038,
    30.96116505,
]

x = [1000 * i for i in 交联剂比例]  # x = 交联剂比例 * 1000
y = 吸水倍率
coefs = np.polyfit(x, y, 2)
x_fit = np.linspace(int(min(x)), int(max(x)), 100)
y_pred = np.polyval(coefs, x_fit)


def plot_交联剂比例(*args):
    plt.figure()
    # plt.bar(x, y, color='red', width=6)
    x_highlight = [x[arg - 1] for arg in args]
    y_highlight = [y[arg - 1] for arg in args]
    plt.bar(x_highlight, y_highlight, color='red', width=6)
    x_remain = list(filter(lambda x: x not in x_highlight, x))
    y_remain = list(filter(lambda y: y not in y_highlight, y))
    plt.bar(x_remain, y_remain, color='pink', width=6)

    plt.plot(x, y, color='blue', marker='*', linestyle='--', linewidth=1, label='实际数据')
    plt.plot(
        x_fit,
        y_pred,
        color='green',
        linestyle='-',
        linewidth=1,
        alpha=0.6,
        label='拟合曲线(二次)',
    )
    plt.xlabel('交联剂比例/‰')
    plt.ylabel('吸水倍率')
    plt.title('交联剂比例(‰)-吸水倍率')
    for i in range(len(x)):
        plt.text(x[i], y[i] + 2, f'{y[i]:.2f}', ha='center', va='bottom', fontsize=9, color='green')
    plt.text(x[2], y[2] - 10, '拟合偏差较大', ha='center', va='bottom', fontsize=9, color='purple')
    plt.plot(x[2], y[2], '*', color='purple')
    plt.figtext(
        0.5,
        -0.07,
        f"""丙烯酸质量 = 5g | 引发剂比例 = 1.6% | 中和度 = 50%
        取样偏差 < 5%
        拟合曲线: y={coefs[0]:.2f}x^2+{coefs[1]:.2f}x+{coefs[2]:.2f}""",
        ha='center',
        fontsize=10,
        color='gray',
    )
    plt.grid(True, linestyle='--')
    plt.legend()
    plt.show()


# plot_交联剂比例(1, 2, 3)
# plot_交联剂比例(4, 5)
# plot_交联剂比例(6, 7)
plot_交联剂比例()
```
