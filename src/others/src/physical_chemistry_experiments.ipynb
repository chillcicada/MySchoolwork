```python
# 参数
flag_water = True  # 水为液态


# data 燃烧热
D = 2000  # g 水的质量
C_water = 4.18  # J/gK 水的比热容
# we don't need the data above
# 比热容
Cpm_water = 75.291  # J/(mol·K) 水的摩尔恒压热容
Cpm_carbondioxide = 37.11  # J/(mol·K) 二氧化碳的摩尔恒压热容
Cpm_oxigen = 29.36  # J/(mol·K) 氧气的摩尔恒压热容
Cpm_苯甲酸 = 146.8  # J/(mol·K) 苯甲酸的摩尔恒压热容
Cpm_萘 = 165.7  # J/(mol·K) 萘的摩尔恒压热容

R = 8.314  # J/(mol·K) 气体常数

m_棉线 = 0.0170  # g 棉线的质量
Qv_棉线 = -16736  # J/gK 棉线的恒容燃烧热

m_点火丝 = 0.0093  # g 点火丝的质量
m_点火丝剩余 = 0.0060  # g 点火丝的剩余质量
Qv_点火丝 = -3243  # J/gK 点火丝的恒容燃烧热

m = 0.7982  # g 总质量
m_苯甲酸 = m - m_棉线 - m_点火丝  # g 苯甲酸的质量
Mr_苯甲酸 = 122.12  # g/mol 苯甲酸的摩尔质量
T = (21.603 + 23.6559) / 2 + 273.15  # K 初始温度
Qp_苯甲酸 = -26460  # J/g 苯甲酸的恒压燃烧热
Qv_苯甲酸 = Qp_苯甲酸 - R * T * 1 / Mr_苯甲酸 * (-0.5 if flag_water else 2.5)  # J/g 苯

delta_T = 1.991  # K 温度变化
print(f'delta_T = {delta_T:.3f} K')
delta_TT = 25 + 273.15 - T  # Kirchhoff's law

Q_点火丝理论 = m_点火丝 * Qv_点火丝
print(f'Q_点火丝理论 = {Q_点火丝理论:.2f} J')

Q_点火丝实际 = (m_点火丝 - m_点火丝剩余) * Qv_点火丝
print(f'Q_点火丝实际 = {Q_点火丝实际:.2f} J')

Q_棉线 = m_棉线 * Qv_棉线
print(f'Q_棉线 = {Q_棉线:.2f} J')

Q_苯甲酸 = m_苯甲酸 * Qv_苯甲酸 - m_苯甲酸 / Mr_苯甲酸 * Cpm_苯甲酸 * delta_TT
print(f'Q_苯甲酸 = {Q_苯甲酸:.2f} J')

K = (Q_苯甲酸 - Q_点火丝实际 - Q_棉线) / delta_T
print(f'K = {K:.2f} J/K')
```

```python
m_棉线 = 0.0184  # g 棉线的质量
m_点火丝 = 0.0094  # g 点火丝的质量
m_点火丝剩余 = 0.0052  # g 点火丝的剩余质量
m = 0.6844  # g 总质量
m_萘 = m - m_棉线 - m_点火丝  # g 萘的质量
T = (21.7221 + 24.3713) / 2 + 273.15  # K 初始温度
Mr_萘 = 128.174  # g/mol 萘的摩尔质量
delta_T = 2.651 - 0.037  # K 温度变化
print(f'delta_T = {delta_T:.3f} K')
delta_TT = 25 + 273.15 - T  # Kirchhoff's law

Q_点火丝理论 = m_点火丝 * Qv_点火丝
print(f'Q_点火丝理论 = {Q_点火丝理论:.2f} J')

Q_点火丝实际 = (m_点火丝 - m_点火丝剩余) * Qv_点火丝
print(f'Q_点火丝实际 = {Q_点火丝实际:.2f} J')

Q_棉线 = m_棉线 * Qv_棉线
print(f'Q_棉线 = {Q_棉线:.2f} J')

Qv_萘 = Mr_萘 * (K * delta_T - Q_点火丝实际 - Q_棉线) / m_萘 + Cpm_萘 * delta_TT
Qp_萘 = Qv_萘 + R * T * (-2 if flag_water else 2)

print(f'Qv_萘 = {Qv_萘:.2f} J/mol = {Qv_萘 / 1000:.2f} kJ/mol')
print(f'Qp_萘 = {Qp_萘:.2f} J/mol = {Qp_萘 / 1000:.2f} kJ/mol')

# error
error = -(Qp_萘 / 1000 + 5153.8) / 5153.8
print(f'error = {error:.2%}')
```
